```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
```

# Calculating the sample mean

```python
late_shipments = pd.read_feather('data/late_shipments.feather')
pd.set_option('display.max_columns', 100)
pd.set_option('display.max_rows', 1000)
```

```python
late_shipments.shape
```

```python
late_shipments.head()
```

```python
late_prop_samp = (late_shipments['late'] == 'Yes').mean()
late_prop_samp
```

```python
# Cool calculating! The proportion of late shipments 
# in the sample is 0.061, or 6.1%.
```

# Calculating a z-score

```python
late_shipments.head()
```

```python
# Hypothesize that the proportion is 6%
late_prop_hyp = 0.06
```

```python
late_shipments_boot_distn = []

for i in range(5000):
    late_shipments_boot_distn.append(
        np.mean(late_shipments.sample(n=1000, replace=True)['late']=="Yes")
    )
```

```python
# Calculate the standard error
std_error = np.std(late_shipments_boot_distn, ddof=1)

# Find z-score of late_prop_samp
z_score = (late_prop_samp - late_prop_hyp) / std_error

z_score
```

# A tail of two z's

## Calculating p-values

```python
late_shipments.head()
```

```python
# Calculate the z-score of late_prop_samp
z_score = (late_prop_samp - late_prop_hyp) / std_error

# Calculate the p-value
p_value = 1 - norm.cdf(z_score, 0, 1)

p_value
```

# Statistically significant other

## Calculating a confidence interval

```python
late_shipments.head()
```

```python
# Calculate 95% confidence interval using quantile method

lower = np.quantile(late_shipments_boot_distn, 0.025)
upper = np.quantile(late_shipments_boot_distn, 0.975)

print(lower,'|', upper)
```
