```python
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('fivethirtyeight')
```

```python
puppies = pd.read_csv('./data/puppies.csv')
```

```python
puppies.head()
```

```python
plt.hist(puppies.weight)
plt.show()
```

```python
plt.hist(puppies.weight, bins=50)
plt.show()
```

```python
plt.hist(puppies.weight, bins=50, range=(5, 35))

plt.xlabel('Puppy Weight (lbs)')
plt.ylabel('Number of Puppies')

plt.show()
```

```python
gravel = pd.read_csv('./data/gravel.csv')
```

```python
gravel.head()
```

```python
plt.hist(gravel.radius)
plt.show()
```

```python
plt.hist(gravel.radius, range=(2,8), bins=40)

plt.show()
```

```python
plt.hist(gravel.radius, range=(2,8), bins=40, density=1)

plt.show()
```

```python
plt.hist(gravel.radius, range=(2,8), bins=40, density=1)

plt.xlabel('Gravel Radius (mm)')
plt.ylabel('Frequency')
plt.title('Sample from Shoeprint')

plt.show()
```
