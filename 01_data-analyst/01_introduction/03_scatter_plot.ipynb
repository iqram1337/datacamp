```python
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('fivethirtyeight')
```

```python
cellphone = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vQYe0Lsydx6HyVe4FBHjBsLxvAsjii-qobZ-oGGZryFJWRFjemjs43BwZQONRAhLCawSUBkBDI__-Xw/pub?gid=65169422&single=true&output=csv')
```

```python
cellphone.head()
```

```python
plt.scatter(cellphone.x, cellphone.y)

plt.ylabel('Latitude')
plt.xlabel('Longitude')
plt.show()
```

```python
plt.scatter(cellphone.x, cellphone.y, color='red', marker='s')

plt.ylabel('Latitude')
plt.xlabel('Longitude')
plt.show()
```

```python
plt.scatter(cellphone.x, cellphone.y, color='red', marker='s', alpha=0.1)

plt.ylabel('Latitude')
plt.xlabel('Longitude')
plt.show()
```
