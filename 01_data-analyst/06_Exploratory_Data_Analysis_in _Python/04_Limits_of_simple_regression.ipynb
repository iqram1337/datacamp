```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import linregress
import statsmodels.formula.api as smf
```

```python
brfss = pd.read_hdf('data/brfss.hdf5')
```

```python
brfss.head()
```

```python
brfss.shape
```

# Using StatsModels

```python
subset = brfss.dropna(subset=['INCOME2', '_VEGESU1'])
xs = subset['INCOME2']
ys = subset['_VEGESU1']INCOME2

subset.head()
```

```python
xs.head()
```

```python
res = linregress(xs, ys)
res
```

```python
results = smf.ols('INCOME2 ~ _VEGESU1', data=subset).fit()

results.params
```

# Multiple regression

```python
gss = pd.read_hdf('data/gss.hdf5')
```

```python
gss.head()
```

## Plot income and education

```python
grouped = gss.groupby('educ')
grouped.head()
```

```python
mean_income_by_educ = grouped['realinc'].mean()
mean_income_by_educ.head()
```

```python
plt.plot(mean_income_by_educ, 'o', alpha=0.5)

plt.show()
```

```python
plt.plot(mean_income_by_educ, 'o', alpha=0.5)

plt.xlabel('Education (years)')
plt.ylabel('Income (1986 $)')
plt.show()
```

## Non-linear model of education

```python
gss['educ2'] = gss['educ']**2
gss['age2'] = gss['age']**2
```

```python
gss.head()
```

```python
results = smf.ols('realinc ~ educ + educ2 + age + age2', data=gss).fit()
results.params
```

# Visualizing regression results

### Making predictions

```python
gss.head()
```

```python
results = smf.ols('realinc ~ age + educ + age2 + educ2', data=gss).fit()
results.params
```

```python
df = pd.DataFrame()
df.head()
```

```python
df['educ'] = np.linspace(0, 20)
df
```

```python
df['age'] = 30
df.head()
```

```python
df['educ2'] = df['educ']**2
df['age2'] = df['age']**2
```

```python
df.head()
```

```python
results.params
```

```python
pred = results.predict(df)
pred
```

### Visualizing predictions

```python
plt.clf()
grouped = gss.groupby('educ')
mean_income_by_educ = grouped['realinc'].mean()
mean_income_by_educ.head()
```

```python
plt.plot(mean_income_by_educ, 'o', alpha=0.5)

plt.xlabel('Education (years)')
plt.ylabel('Income (1986 $)')
plt.show()
```

```python
pred = results.predict(df)
pred.head()
```

```python
plt.plot(df['educ'], pred, label='Age 30')

plt.legend()
plt.xlabel('Education (years)')
plt.ylabel('Income (1986 $)')
plt.show()
```

```python
# Plot mean income in each age group
plt.clf()
grouped = gss.groupby('educ')
mean_income_by_educ = grouped['realinc'].mean()
plt.plot(mean_income_by_educ, 'o', alpha=0.5)

# Plot the predictions
pred = results.predict(df)
plt.plot(df['educ'], pred, label='Age 30')

# Label axes
plt.xlabel('Education (years)')
plt.ylabel('Income (1986 $)')

plt.legend()
plt.show()
```
