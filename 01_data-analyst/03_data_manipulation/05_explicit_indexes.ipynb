# Pivot Table

```python
temperatures = pd.read_csv("./data/temperatures.csv", index_col=0)
temperatures['date'] = pd.to_datetime(temperatures['date'])

temperatures.head()
```

```python
temperatures["date"].dt.month.head()
```

```python
temperatures["date"].dt.day.head()
```

```python
temperatures["year"] = temperatures["date"].dt.year
```

```python
temperatures.head()
```

```python
temp_by_country_city_vs_year = temperatures.pivot_table(index=["country","city"], columns="year", values="avg_temp_c")
temp_by_country_city_vs_year.head()
```

```python
temp_by_country_city_vs_year.loc["Egypt":"India"]
```

```python
temp_by_country_city_vs_year.loc[("Egypt", "Cairo"):("India", "Delhi")]
```

```python
temp_by_country_city_vs_year.loc[("Egypt", "Cairo"):("India", "Delhi"), "2005":"2010"]
```

# Calculating on a pivot table

```python
temp_by_country_city_vs_year.head()
```

```python
mean_temp_by_year = temp_by_country_city_vs_year.mean()
mean_temp_by_year
```

```python
mean_temp_by_year[mean_temp_by_year == mean_temp_by_year.max()]
```

```python
mean_temp_by_city = temp_by_country_city_vs_year.mean(axis="columns")
mean_temp_by_city
```

```python
mean_temp_by_city[mean_temp_by_city == mean_temp_by_city.min()]
```
