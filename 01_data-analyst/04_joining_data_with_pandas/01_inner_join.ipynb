# One-to-many relationships

```python
biz_owners = pd.read_pickle("./data/business_owners.p")
licenses = pd.read_pickle("./data/licenses.p")
```

```python
biz_owners.head()
```

```python
licenses.head()
```

```python
licenses_owners = licenses.merge(right=biz_owners, on="account")
licenses_owners.head()
```

```python
licenses_owners.shape
```

```python
counted_df = licenses_owners.groupby("title").agg({"account":"count"})
counted_df
```

```python
sorted_df = counted_df.sort_values("account", ascending=False)
sorted_df
```

# Merging multiple DataFrames

```python
cal = pd.read_pickle("./data/calendar.p")
ridership = pd.read_pickle("./data/ridership.p")
stations = pd.read_pickle("./data/stations.p")
```

```python
cal.head()
```

```python
ridership.head()
```

```python
stations.head()
```

```python
ridership_cal = ridership.merge(right=cal, on=["year", "month", "day"])
ridership_cal
```

```python
ridership_cal_stations = ridership.merge(right=cal, on=["year", "month", "day"]).merge(right=stations, on=["station_id"])
ridership_cal_stations
```

```python
filter_criteria = ((ridership_cal_stations['month'] == 7) 
                   & (ridership_cal_stations['day_type'] == "Weekday") 
                   & (ridership_cal_stations['station_name'] == "Wilson"))
```

```python
ridership_cal_stations.loc[filter_criteria, "rides"].sum()
```

# Three table merge

```python
zip_demo = pd.read_pickle("./data/zip_demo.p")
wards = pd.read_csv("./data/wards.csv")
licenses = pd.read_pickle("./data/licenses.p")
```

```python
zip_demo.head()
```

```python
wards.head()
```

```python
licenses.head()
```

```python
wards["ward"] = wards["ward"].astype(int)
licenses["ward"] = licenses["ward"].astype(int)

licenses_zip_ward = licenses.merge(right=zip_demo, on="zip").merge(right=wards, on="ward")

licenses_zip_ward.head()
```

```python
licenses_zip_ward.groupby("alderman").agg({"income":"median"})
```

# One-to-many merge with multiple tables

```python
land_use = pd.read_pickle("./data/land_use.p")
```

```python
land_use = land_use.astype(int)
wards["ward"] = wards["ward"].astype(int)
licenses["ward"] = licenses["ward"].astype(int)
census["ward"] = census["ward"].astype(int)
```

```python
land_use.head()
```

```python
census.head()
```

```python
licenses.head()
```

```python
land_cen_lic = land_use.merge(right=census, on="ward").merge(right=licenses, on="ward", suffixes=["_cen", "_lic"])
land_cen_lic.head()
```

```python
pop_vac_lic = land_cen_lic.groupby(["ward", "pop_2010", "vacant"]).agg({"account":"count"})
pop_vac_lic.head()
```

```python
pop_vac_lic = land_cen_lic.groupby(["ward", "pop_2010", "vacant"], as_index=False).agg({"account":"count"})
pop_vac_lic.head()
```

```python
sorted_pop_vac_lic = pop_vac_lic.sort_values(by=["vacant","account","pop_2010"], ascending=[False, True, True])
sorted_pop_vac_lic.head()
```
