# Customizing bar plots

```python
student_data = pd.read_csv('data/student_data.csv', index_col=0)
pd.set_option('display.max_columns', 100)
```

```python
student_data.head()
```

```python
sns.catplot(x='study_time', y='G3', data=student_data, kind='bar')

plt.show()
```

```python
category_order = ["<2 hours", 
                  "2 to 5 hours", 
                  "5 to 10 hours", 
                  ">10 hours"]

sns.catplot(x='study_time', y='G3', data=student_data, kind='bar', order=category_order)

plt.show()
```

```python
category_order = ["<2 hours", 
                  "2 to 5 hours", 
                  "5 to 10 hours", 
                  ">10 hours"]

sns.catplot(x='study_time', y='G3', data=student_data, kind='bar', order=category_order, ci=None)

plt.show()
```

# Box plot

```python
student_data.head()
```

```python
study_time_order = ["<2 hours", "2 to 5 hours", 
                    "5 to 10 hours", ">10 hours"]

sns.catplot(x='study_time', y='G3', data=student_data, kind='box', order=study_time_order)

plt.show()
```

# Adjusting the whiskers

```python
student_data.head()
```

```python
sns.catplot(x="romantic", y="G3", data=student_data, kind="box")

plt.show()
```

```python
sns.catplot(x="romantic", y="G3", data=student_data, kind="box", whis=0.5)

plt.show()
```

```python
sns.catplot(x="romantic", y="G3", data=student_data, kind="box", whis=[5, 95])

plt.show()
```

```python
sns.catplot(x="romantic", y="G3", data=student_data, kind="box", whis=[0, 100])

plt.show()
```

# Customizing point plots

```python
sns.catplot(x='famrel', y='absences', data=student_data, kind='point')
plt.show()
```

# Point plots with subgroups
