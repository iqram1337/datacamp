```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Customizing point plots

```python
survey_data = pd.read_csv("data/survey_data3.csv")
pd.set_option('display.max_columns', 100)

survey_data.rename(columns={"Parents' advice": "Parents Advice"}, inplace=True)
```

```python
survey_data.head()
```

```python
sns.catplot(x="Parents Advice", data=survey_data, kind='count')
plt.show()
```

```python
category_order = ["Never", "Rarely", "Sometimes", "Often", "Always"]

sns.set_style('whitegrid')

sns.catplot(x="Parents Advice", data=survey_data, kind='count', order=category_order)
plt.show()
```

```python
category_order = ["Never", "Rarely", "Sometimes", "Often", "Always"]

sns.set_style('whitegrid')
sns.set_palette('Purples')

sns.catplot(x="Parents Advice", data=survey_data, kind='count', order=category_order)
plt.show()
```

```python
category_order = ["Never", "Rarely", "Sometimes", "Often", "Always"]

sns.set_style('whitegrid')
sns.set_palette('RdBu')

sns.catplot(x="Parents Advice", data=survey_data, kind='count', order=category_order)
plt.show()
```

## FacetGrids vs. AxesSubplots

```python
mpg = sns.load_dataset('mpg')
```

```python
mpg.head()
```

```python
g = sns.relplot(x="weight", y="horsepower", data=mpg, kind="scatter")

plt.show()
```

## Adding a title and axis labels

```python
mpg_mean = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vQL_SK_cXnLjskBmohM5wuw8cTtKGGOfUF9vZZQIPu_QLffOmFq0dwmSa46ChvNV9WN4SRnHFrtktF0/pub?gid=0&single=true&output=csv')
mpg_mean.head()
```

```python
g = sns.lineplot(x="model_year", y="mpg_mean", data=mpg_mean, hue="origin")

plt.show()
```

```python
g = sns.lineplot(x="model_year", y="mpg_mean", data=mpg_mean, hue="origin")


g.set_title("Average MPG Over Time")
g.set(xlabel="Car Model Year", ylabel="Average MPG")

plt.show()
```

## Rotating x-tick labels

```python
sns.catplot(x="origin", y="acceleration", data=mpg, kind="point", join=False, capsize=0.1)

plt.show()
```

```python
sns.catplot(x="origin", y="acceleration", data=mpg, kind="point", join=False, capsize=0.1)

plt.xticks(rotation=90)
plt.show()
```

## Bar plot with subgroups and subplots

```python
survey_data = pd.read_csv('data/survey_data5.csv')
survey_data.head()
```

```python
g = sns.catplot(x="Village - town", y="Likes Techno", data=survey_data, kind="bar", col='Gender')

plt.show()
```

```python
sns.set_style('dark')

g = sns.catplot(x="Village - town", y="Likes Techno", data=survey_data, kind="bar", col='Gender')

g.fig.suptitle("Percentage of Young People Who Like Techno", y=1.02)
g.set(xlabel="Location of Residence", ylabel="% Who Like Techno")

plt.show()
```
