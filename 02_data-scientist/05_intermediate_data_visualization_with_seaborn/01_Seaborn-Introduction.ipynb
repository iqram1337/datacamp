## Reading a csv file

```python
# import all modules
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

grant_file = ("data/schoolimprovement2010grants.csv")

# Read in the DataFrame
df = pd.read_csv(grant_file, index_col=0)
```

```python
df.head()
```

## Comparing a histogram and displot

```python
# Display pandas histogram
df['Award_Amount'].plot.hist()
plt.show()
```

```python
# Display a Seaborn histplot
sns.histplot(df['Award_Amount'])
plt.show()
```

```python
# Display a Seaborn displot
sns.displot(df['Award_Amount'])
plt.show()
```

# Using the distribution plot

## Plot a histogram

```python
# Create a displot
sns.displot(df['Award_Amount'], bins=20)

# Display the plot
plt.show()
```

## Rug plot and kde shading

```python
# Create a displot of the Award Amount
sns.displot(df['Award_Amount'], kind='kde', rug=True, fill=True)

# Plot the results
plt.show()
```

# Regression Plots in Seaborn

## Create a regression plot

```python
FiveThirtyEight = "data/insurance_premiums.csv"

df = pd.read_csv(FiveThirtyEight)
```

```python
df.head()
```

```python
# Create a regression plot of premiums vs. insurance_losses
sns.regplot(data=df, x="insurance_losses", y="premiums")

# Display the plot
plt.show()
```

```python
# Create an lmplot of premiums vs. insurance_losses
sns.lmplot(data=df, x="insurance_losses", y="premiums")

# Display the second plot
plt.show()
```

## Plotting multiple variables

```python
# Create a regression plot using hue
sns.lmplot(data=df,
           x="insurance_losses",
           y="premiums",
           hue="Region")

# Show the results
plt.show()
```

## Facetting multiple regressions

```python
sns.lmplot(data=df,
           x="insurance_losses",
           y="premiums",
           col="Region")

# Show the results
plt.show()
```
